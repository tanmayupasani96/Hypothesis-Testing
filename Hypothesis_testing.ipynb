```python
sample_df = df.sample(35, random_state=10)
sample_df
```

```python
salmean = sample_df['2019_Average_farm_size(acres)'].mean()
salstd = sample_df['2019_Average_farm_size(acres)'].std()
print(salmean, salstd)
```

```python
x = salmean   #sample mean
mu = 446.92    #Population mean
S = salstd    # Standard deviation
n=35          # sample size
alpha = 0.05 #level of significance
```

```python
import math
from scipy.stats import norm
z_stat = (x-mu)/(S/math.sqrt(n))

p_val=norm.sf(abs(z_stat))
print(p_val)
```

Since our p value is for one tailed right test is greater than alpha, we do NOT reject null hypothesis. If P value is less thanm 0.05 we reject null hypothesis. If P value is greater than 0.05 we reject the alternate hypothesis or accept the null hypothesis

```python
df = pd.DataFrame({"gender":['M','M','M','F','F','M','F','F','F','M','F'],
                   "stand":['a','a','b','a','a','b','b','b','a','b','a'],
                   "marks":[20,13,13,20,34,21,60,32,36,32,21]
                   })
```

```python
df
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
model = ols('marks ~ C(gender) + C(stand) + C(gender):C(stand)', data = df).fit()  #check dependance of marks columns on other columns
annova_table = sm.stats.anova_lm(model, type = 2)
print(annova_table)
```

```python
#The last value i.e. 0.377 is greater than 0.05 which is greater than 0.05 so we accept null hypo
#If greater than 0.05 we accept null hypo
#Null hypo: There is no association
#Alternative hypo: There is association
```

```python
if annova_table['PR(>F)'][2]<0.05:
  print('Null hypo fails and there is significant association between marks and gender and standard')
else:
  print("Null Hypo is accepted and and there is no significant association between marks and gender and standard")
```
